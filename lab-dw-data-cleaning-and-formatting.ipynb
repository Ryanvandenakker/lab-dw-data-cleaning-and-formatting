## Exercise 1: Cleaning Column Names

To ensure consistency and ease of use, standardize the column names of the dataframe. Start by taking a first look at the dataframe and identifying any column names that need to be modified. Use appropriate naming conventions and make sure that column names are descriptive and informative.

*Hint*:
- *Column names should be in lower case*
- *White spaces in column names should be replaced by `_`*
- *`st` could be replaced for `state`*

```python
import pandas as pd
insurance_df = pd.read_csv("https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file1.csv")
insurance_df.head()
```

```python
#Column names in lower case
insurance_df.rename(columns={col:col.replace(" ", "_").lower() for col in insurance_df.columns}, inplace = True)
```

```python
#Replace "st" for "state"
insurance_df.rename(columns = {"st": "state"}, inplace = True)
display(insurance_df)
```

## Exercise 2: Cleaning invalid Values

The dataset contains columns with inconsistent and incorrect values that could affect the accuracy of our analysis. Therefore, we need to clean these columns to ensure that they only contain valid data.

Note that this exercise will focus only on cleaning inconsistent values and will not involve handling null values (NaN or None).

*Hint*:
- *Gender column contains various inconsistent values such as "F", "M", "Femal", "Male", "female", which need to be standardized, for example, to "M" and "F".*
- *State abbreviations be can replaced with its full name, for example "AZ": "Arizona", "Cali": "California", "WA": "Washington"*
- *In education, "Bachelors" could be replaced by "Bachelor"*
- *In Customer Lifetime Value, delete the `%` character*
- *In vehicle class, "Sports Car", "Luxury SUV" and "Luxury Car" could be replaced by "Luxury"*

```python
state_values = insurance_df['state'].value_counts()
print(state_values)
print()
edu = insurance_df["education"].value_counts()
print(edu)
print()
car_class = insurance_df["vehicle_class"].value_counts()
print(car_class)
```

```python
insurance_df["gender"].unique()
```

```python
#standardize 'gender' column
insurance_df["gender"] = insurance_df["gender"].fillna("na").apply(lambda x: 'F' if x.lower() in ['f', 'female', 'femal'] else 'M' if x.lower() in ['m', 'male'] else x)
insurance_df
```

```python
#Return the fillna string ('na') back to the float 'NaN'
import numpy as np
insurance_df["gender"] = insurance_df["gender"].replace('na', np.nan)
```

```python
#State abbreviations
abbreviations = {"Oregon": "OR", "California": "CA", "Arizona": "AZ", "Nevada": "NV", "Washington": "WA", "Cali":"CA", "AZ":"AZ", "WA":"WA"}
insurance_df["state"] = insurance_df["state"].map(abbreviations)
insurance_df
```

```python
insurance_df["state"].unique()
```

```python
#Bachelors replaced by Bachelor
insurance_df["education"] = insurance_df["education"].apply(lambda x: "Bachelor" if x == "Bachelors" else x)
insurance_df
```

```python
insurance_df["education"].unique()
```

```python
#Remove the % from the "Customer Lifetime Value" column
insurance_df["customer_lifetime_value"] = insurance_df["customer_lifetime_value"].astype(str).apply(lambda x: x.replace("%", ""))
insurance_df
```

```python
#"Sports Car", "Luxury SUV" and "Luxury Car" replaced by "Luxury"
vehicle_mapping = {"Sports Car":"Luxury", "Luxury SUV":"Luxury", "Luxury Car":"Luxury"}
insurance_df["vehicle_class"] = insurance_df["vehicle_class"].replace(vehicle_mapping)
insurance_df
```
